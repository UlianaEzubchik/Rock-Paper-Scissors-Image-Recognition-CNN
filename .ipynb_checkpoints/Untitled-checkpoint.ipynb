```python
import tensorflow as tf
import os
import urllib.request
import zipfile
import tensorflow as tf
import cv2
import mediapipe as mp
import time
import numpy as np

# Load the saved model
loaded_model = tf.keras.models.load_model("rps_model.h5")

cap = cv2.VideoCapture(0)

mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils

pTime = 0
cTime = 0

while True:
    success, img = cap.read()
    if not success:
        print("Failed to capture frame from camera.")
        break

    # Convert the image to RGB format
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Process the hand landmarks using MediaPipe Hands
    results = hands.process(imgRGB)

    # Check if hand landmarks are detected
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            # Extract hand landmarks and reshape to match the input shape of the model
            hand_landmarks = np.array([[lm.x, lm.y, lm.z] for lm in handLms.landmark])
            hand_landmarks = np.expand_dims(hand_landmarks, axis=0)  # Add batch dimension
            hand_landmarks = np.resize(hand_landmarks, (1, 150, 150, 3))  # Resize to match input shape of the model

            # Predict gesture using the loaded model
            prediction = loaded_model.predict(hand_landmarks)
            gesture = np.argmax(prediction)  # Get the index of the highest probability

            # Convert the gesture index to corresponding text
            gesture_text = ["Rock", "Paper", "Scissors"][gesture]

            # Display prediction on the screen
            cv2.putText(img, gesture_text, (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

            # Draw landmarks and connections
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)

    # Calculate and display FPS
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    cv2.putText(img, f'FPS: {int(fps)}', (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    # Display the processed image
    cv2.imshow("Image", img)

    # Check for exit key (ESC)
    if cv2.waitKey(1) == 27:
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
